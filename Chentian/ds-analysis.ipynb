```python
import pandas as pd
import os 
```

```python
path = '../data/vulnerabilitydataset/'
sub_dirs = os.listdir(path)
sub_dirs
```

```python
ldf = pd.DataFrame(columns = ['class', 'neg_length','pos_length','neg_size','pos_size','size'])
ldf['class'] = sub_dirs[1:-5]
ldf = ldf.set_index('class')
ldf
```

```python
for sub_dir in sub_dirs:
    if sub_dir[0:3] == "CWE":
        sub_path = path+sub_dir+"\\"
        files = os.listdir(sub_path)
        size = len(files)
        neg_len = 0
        pos_len = 0
        neg_size = 0
        pos_size = 0
        for file_id in files:
            if file_id[-5] == '0':
                file = open(sub_path+file_id,"r")
                neg_size += 1
                neg_len += len(file.read())
            elif file_id[-5] == '1':
                file = open(sub_path+file_id,"r")
                pos_size+=1
                pos_len += len(file.read())
        neg_len = neg_len/neg_size if neg_size != 0 else 0
        pos_len = pos_len/pos_size if pos_size != 0 else 0
        ldf.at[sub_dir,'neg_length'] = neg_len
        ldf.at[sub_dir,'pos_length'] = pos_len
        ldf.at[sub_dir,'neg_size'] = neg_size
        ldf.at[sub_dir,'pos_size'] = pos_size
        ldf.at[sub_dir,'size'] = size
```

```python
pd.set_option('display.max_rows', None)
ldf
```

```python
import matplotlib.pyplot as plt
import numpy as np

objects = ldf.index
y_pos = np.arange(len(objects))
neg_length=ldf['neg_length']
pos_length=ldf['pos_length']
plt.figure(figsize=(50,50))
plt.bar(y_pos, neg_length, align='center', alpha=0.5)
plt.bar(y_pos, pos_length, align='center', alpha=0.5)

plt.xticks(y_pos, objects)
# plt.ylabel('')
# plt.title('')

plt.show()
```
